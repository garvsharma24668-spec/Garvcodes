```python
def dfs(start, goal, graph, visited=None):
   
    if visited is None:
        visited = set()

    print(start)

    if start == goal:
        return True

    visited.add(start)

    for child in graph.get(start, []): 
        if child not in visited:
            if dfs(child, goal, graph, visited):
                return True

    return False


graph = {
    "A": ["B", "C"],
    "B": ["A", "C"],
    "C": ["A", "B", "D"],
    "D": ["C", "E"],
    "E": ["D"]
}

found = dfs("A", "D", graph)
print("Goal found:", found)
```

```python
from collections import deque

def bfs(start, goal, graph):
    queue = deque([[start]])
    visited = set([start])

    while queue:
        path = queue.popleft()  
        node = path[-1]        
        if node == goal:
            return path

        for child in graph.get(node, []):
            if child not in visited:
                visited.add(child)
                new_path = path + [child]
                queue.append(new_path)

    return None 
graph = {
    "A": ["B", "C"],
    "B": ["A", "C"],
    "C": ["A", "B", "D"],
    "D": ["C", "E"],
    "E": ["D"]
}
found = bfs("A", "D", graph)
print("Goal found:", found)
```

```python
graph = {
    "A": ["B", "C"],
    "B": ["A", "D"],
    "C": ["A", "E", "F"],
    "D": ["B"],
    "E": ["C"],
    "F":["C"]
}
found = dfs("A", "F", graph)
print("Goal found:", found)
found = bfs("A", "F", graph)
print("Goal found:", found)
```
